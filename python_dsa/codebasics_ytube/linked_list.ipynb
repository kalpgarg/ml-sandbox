```python
class Node:
    def __init__(self, data=None, next=None) -> None:
        self.data = data
        self.next = next
```

```python
class LinkedList:
    def __init__(self) -> None:
        self.head = None
    
    def insert_at_beginning(self,data) -> None:
        """
        Insert data at the beginning
        """
        node = Node(data, self.head)
        self.head = node

    def insert_at_end(self, data) -> None:
        """
        Insert data at the end
        """
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data,None)

    def insert_values(self, data_list) -> None:
        """
        Insert data list 
        """
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    
    def get_length(self) -> int:
        """
        Get the length of Linked list
        """
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        
        return count
    
    def remove_at(self, index) -> None:
        """
        Remove element at particular index
        """
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                break
            itr = itr.next
            count += 1

    def insert_at(self, index, data) -> None:
        """
        Insert element at particular index
        """
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                node = Node(data,itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
    
    def insert_after_value(self, data_after, data_to_insert) -> None:
        """
        Search for first occurence of data and insert data after that occurence
        """
        count = 0
        itr = self.head
        while itr:
            if itr.data == data_after:
                node = Node(data_to_insert, itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
        if count == self.get_length():
            raise Exception(f"Data value: {data_after} doesn't exist in the Linked list")
        
    def remove_by_value(self, data_to_remove) -> None:
        """
        Remove first occurence of given data from linked list
        """
        count = 0
        itr = self.head
        while itr:
            if itr.next.data == data_to_remove:
                itr.next = itr.next.next
                break
            itr = itr.next
            count += 1
        if count == self.get_length():
            raise Exception(f"Data value: {data_to_remove} doesn't exist in the Linked list")


    def print(self) -> None:
        """
        Print the Linked List
        """
        if self.head is None:
            print("Linked List is empty")
            return 
        itr = self.head
        llstr = ''

        while itr:
            llstr += str(itr.data) + '---->'
            itr = itr.next

        print(llstr)
```

```python
ll = LinkedList()
ll.insert_at_beginning(34)
ll.insert_at_beginning(56)
ll.insert_at_end(55)
ll.print()
ll.get_length()
ll.remove_at(1)
ll.print()
ll.insert_at(1, 34)
ll.insert_at(1, 43)
ll.print()
ll.insert_after_value(56, 78)
ll.print()
ll.remove_by_value(55)
ll.print()
ll.remove_by_value(43)
ll.print()
```
